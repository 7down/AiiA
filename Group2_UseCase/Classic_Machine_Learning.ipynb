```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import pipeline
import joblib
import os

# Load the cleaned CSV file
df = pd.read_csv("df_filtered.csv")

# Step 1: Add sentiment analysis
print("Starting sentiment analysis...")

# Use Hugging Face's Transformers Pipeline for sentiment analysis
sentiment_analyzer = pipeline("sentiment-analysis")
# Ensure all entries are strings
df['Filtered_Content'] = df['Filtered_Content'].fillna("").astype(str)

# Function to compute sentiment
def compute_sentiment(text):
    if isinstance(text, str) and len(text) > 0:
        result = sentiment_analyzer(text[:512])[0]  # Limit text to 512 characters
        return result['label'], result['score']
    else:
        return "NEUTRAL", 0.0  # Default values for invalid entries

# Calculate sentiment and scores
df['Sentiment'], df['Sentiment_Score'] = zip(*df['Filtered_Content'].apply(compute_sentiment))

# Save the results to a file
df.to_csv("sentiment_results.csv", index=False)
print("Sentiment analysis completed and saved.")

# Step 2: TF-IDF Vectorization
print("Starting TF-IDF vectorization...")
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Limit to 1000 features for better performance
X_tfidf = tfidf_vectorizer.fit_transform(df['Filtered_Content'])
y = df['NegoOutcomeLabel']

# Save the TF-IDF vectorizer
joblib.dump(tfidf_vectorizer, "tfidf_vectorizer.pkl")

# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Step 4: Training a more complex model
print("Starting training of a RandomForest classifier...")
rf_classifier = RandomForestClassifier(random_state=42, n_estimators=100)
rf_classifier.fit(X_train, y_train)

# Predictions
y_pred = rf_classifier.predict(X_test)

# Evaluation
print("Model results:")
print("Model Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Save the model
joblib.dump(rf_classifier, "random_forest_model.pkl")

# Step 5: Save the results
results = {
    "accuracy": accuracy_score(y_test, y_pred),
    "classification_report": classification_report(y_test, y_pred, output_dict=True)
}
results_df = pd.DataFrame(results)
results_df.to_csv("model_results.csv", index=False)

print("All results saved.")

# Document the top TF-IDF features
feature_importances = rf_classifier.feature_importances_
feature_names = tfidf_vectorizer.get_feature_names_out()
important_features = sorted(zip(feature_importances, feature_names), reverse=True)[:50]
important_features_df = pd.DataFrame(important_features, columns=["Importance", "Feature"])
important_features_df.to_csv("top_tfidf_features.csv", index=False)

print("Top TF-IDF Features saved.")

# Step 6: Correlation between sentiment and success
print("Calculating correlation between sentiment and success...")
sentiment_success = df.groupby(['Sentiment', 'NegoOutcomeLabel']).size().unstack(fill_value=0)
sentiment_success.to_csv("sentiment_success_correlation.csv")

print("Correlation results saved.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
import joblib

# Load and clean the data
df = pd.read_csv("sentiment/sentiment_results.csv")
df['Filtered_Content'] = df['Filtered_Content'].fillna("").astype(str)

# Load saved TF-IDF data and transform the cleaned content
X_tfidf = joblib.load("sentiment/tfidf_vectorizer.pkl").transform(df['Filtered_Content'])
y = df['NegoOutcomeLabel']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Store results and predictions
results = {}
predictions = {}

# Train models and save predictions
print("Training Random Forest...")
rf_model = RandomForestClassifier(random_state=42, n_estimators=100, max_depth=10)
rf_model.fit(X_train, y_train)
predictions['Random Forest'] = rf_model.predict(X_test)
results['Random Forest'] = classification_report(y_test, predictions['Random Forest'], output_dict=True)
joblib.dump(rf_model, "rf_model.pkl")


print("Training XGBoost...")
xgb_model = XGBClassifier(random_state=42, n_estimators=100, max_depth=5, learning_rate=0.1)
xgb_model.fit(X_train, y_train)
predictions['XGBoost'] = xgb_model.predict(X_test)
results['XGBoost'] = classification_report(y_test, predictions['XGBoost'], output_dict=True)
joblib.dump(xgb_model, "xgb_model.pkl")

print("Training LightGBM...")
lgbm_model = LGBMClassifier(random_state=42, n_estimators=100, max_depth=5, learning_rate=0.1)
lgbm_model.fit(X_train, y_train)
predictions['LightGBM'] = lgbm_model.predict(X_test)
results['LightGBM'] = classification_report(y_test, predictions['LightGBM'], output_dict=True)
joblib.dump(lgbm_model, "lgbm_model.pkl")

print("Training SVM...")
svm_model = SVC(kernel='linear', C=1.0, random_state=42)
svm_model.fit(X_train, y_train)
predictions['SVM'] = svm_model.predict(X_test)
results['SVM'] = classification_report(y_test, predictions['SVM'], output_dict=True)
joblib.dump(svm_model, "svm_model.pkl")

print("Training Logistic Regression...")
log_reg = LogisticRegression(random_state=42, max_iter=1000)
log_reg.fit(X_train, y_train)
predictions['Logistic Regression'] = log_reg.predict(X_test)
results['Logistic Regression'] = classification_report(y_test, predictions['Logistic Regression'], output_dict=True)
joblib.dump(log_reg, "logistic_regression_model.pkl")

print("Training MLPClassifier...")
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
mlp_model.fit(X_train, y_train)
predictions['MLPClassifier'] = mlp_model.predict(X_test)
results['MLPClassifier'] = classification_report(y_test, predictions['MLPClassifier'], output_dict=True)
joblib.dump(mlp_model, "mlp_model.pkl")

# Save results
results_df = pd.DataFrame({model: metrics['accuracy'] for model, metrics in results.items()}, index=['Accuracy'])
results_df.to_csv("model_accuracies.csv")

# Save detailed classification reports in a readable format
with open("model_classification_reports.txt", "w") as f:
    for model, metrics in results.items():
        f.write(f"### {model} ###\n")
        f.write(classification_report(y_test, predictions[model]) + "\n\n")

print("All models trained and results saved.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the saved results
try:
    results_df = pd.read_csv("model_accuracies.csv")
except FileNotFoundError:
    print("The file 'model_accuracies.csv' was not found. Please check if the file exists.")
    exit()

classification_reports = {}

# Read classification reports
try:
    with open("model_classification_reports.txt", "r") as f:
        reports = f.read().split("###")[1:]
        for report in reports:
            lines = report.strip().split("\n")
            model_name = lines[0].strip()
            classification_reports[model_name] = "\n".join(lines[1:])
except FileNotFoundError:
    print("The file 'model_classification_reports.txt' was not found. Please check if the file exists.")
    exit()

# 1. Format the results correctly
try:
    results_df_melted = results_df.melt(id_vars="Unnamed: 0", var_name="Model", value_name="Accuracy")
    results_df_melted = results_df_melted[results_df_melted["Unnamed: 0"] == "Accuracy"]
except KeyError:
    print("Error processing 'results_df'. Please check the structure of the file.")
    print(results_df.head())
    exit()

# 2. Plot the accuracy of all models
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df_melted, x="Model", y="Accuracy", palette="viridis")
plt.title("Model Accuracy Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("model_accuracy_comparison.png")
plt.show()

# 3. Compare Precision, Recall, and F1-Score for each model (Weighted Avg)
weighted_metrics = []

import re

# Compare Precision, Recall, and F1-Score for each model (Weighted Avg)
weighted_metrics = []

for model_name, report in classification_reports.items():
    lines = report.split("\n")
    for line in lines:
        if "weighted avg" in line:
            # Use regular expressions to extract numbers from the line
            matches = re.findall(r"\d+\.\d+", line)  # Search for numbers in the format X.XX
            if len(matches) >= 3:  # At least Precision, Recall, and F1-Score must be present
                precision = float(matches[0])
                recall = float(matches[1])
                f1 = float(matches[2])
                weighted_metrics.append((model_name, precision, recall, f1))
            else:
                print(f"Error parsing 'weighted avg' for {model_name}: {line}")

# DataFrame for weighted avg metrics
if weighted_metrics:
    metrics_df = pd.DataFrame(weighted_metrics, columns=["Model", "Precision", "Recall", "F1-Score"])

    # 4. Plot Precision, Recall, and F1-Score
    metrics_df_melted = metrics_df.melt(id_vars="Model", value_vars=["Precision", "Recall", "F1-Score"], 
                                        var_name="Metric", value_name="Value")
    plt.figure(figsize=(12, 8))
    sns.barplot(data=metrics_df_melted, x="Model", y="Value", hue="Metric", palette="Set2")
    plt.title("Weighted Avg Metrics for All Models")
    plt.xlabel("Models")
    plt.ylabel("Metric Value")
    plt.xticks(rotation=45, ha="right")
    plt.legend(title="Metric")
    plt.tight_layout()
    plt.savefig("weighted_avg_metrics_comparison.png")
    plt.show()

    # 5. Insights and Connections
    # Models with the best accuracy and F1-Score
    if not metrics_df.empty:
        best_model = metrics_df.sort_values("F1-Score", ascending=False).iloc[0]

        # Save key results
        key_insights = {
            "Best Model": best_model["Model"],
            "Best Model Precision": best_model["Precision"],
            "Best Model Recall": best_model["Recall"],
            "Best Model F1-Score": best_model["F1-Score"]
        }
        insights_df = pd.DataFrame([key_insights])
        insights_df.to_csv("key_model_insights.csv", index=False)
        
        print("Analysis complete. Plots and insights saved.")
    else:
        print("The DataFrame 'metrics_df' is empty. No insights available.")
else:
    print("No 'weighted avg' values found. Please check the classification reports.")
```
